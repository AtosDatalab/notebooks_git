```python
import json
import urllib.request
import base64
import time
import pandas as pd
from datetime import datetime
import random


def label_retard_pt_stat (row):
    if row['timestamps_eibt']!='None' and row['timestamps_eldt']!='None' and row['timestamps_aibt']!='None' and  row['timestamps_aldt']!='None':
            timestamps_sibt=datetime.strptime(row['timestamps_sibt'], "%Y-%m-%d %H:%M:%S")
            timestamps_aibt=datetime.strptime(row['timestamps_aibt'], "%Y-%m-%d %H:%M:%S")
            if timestamps_aibt<timestamps_sibt:
                return 0
            else:
                return int((timestamps_aibt-timestamps_sibt).seconds/60)
    else:
        return random.randint(0, 30)

def label_retart_terissage(row):
    if row['timestamps_eibt']!='None' and row['timestamps_eldt']!='None' and row['timestamps_aibt']!='None' and  row['timestamps_aldt']!='None':
            timestamps_eldt=datetime.strptime(row['timestamps_eldt'], "%Y-%m-%d %H:%M:%S")
            timestamps_aldt=datetime.strptime(row['timestamps_aldt'], "%Y-%m-%d %H:%M:%S")
            if timestamps_aldt<timestamps_eldt:
                return 0
            else:
                return int((timestamps_aldt-timestamps_eldt).seconds/60)
    else:
        return random.randint(0, 30)


    
def get_year(row):
    
    return  datetime.strptime(row['timestamps_sibt'], "%Y-%m-%d %H:%M:%S").year


   
def get_day(row):
    
    return  datetime.strptime(row['timestamps_sibt'], "%Y-%m-%d %H:%M:%S").day


   
def get_month(row):
    
    return  datetime.strptime(row['timestamps_sibt'], "%Y-%m-%d %H:%M:%S").month


def get_weekday(row):
    
    return  datetime.strptime(row['timestamps_sibt'], "%Y-%m-%d %H:%M:%S").weekday()



def get_hour(row):
    
    return  datetime.strptime(row['timestamps_sibt'], "%Y-%m-%d %H:%M:%S").hour

```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import numpy as np
foresrt_model=RandomForestRegressor(n_estimators=5, random_state=42)
y=df_dummies['retard_pt_stat']
df_dummies=df_dummies.drop('retard_pt_stat',axis=1)
y=np.array(y)
X=df_dummies
X=np.array(X)
train_X,val_X,train_y,val_y=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
foresrt_model.fit(train_X,train_y)
```

```python
flightdelay_pred=foresrt_model.predict(val_X)
errors_randoms=abs(flightdelay_pred-val_y)
```

```python
print(round(np.mean(errors_randoms),6))
```

```python
from sklearn import metrics
print('metrics')
#the best value is 0.0
print('Mean Absolute Error:', metrics.mean_absolute_error(val_y, flightdelay_pred)) 
print('Mean Squared Error:', metrics.mean_squared_error(val_y, flightdelay_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(val_y, flightdelay_pred)))

print('rsquare')
print('r-squardd', foresrt_model.score(train_X,train_y)) 
print('r-squardd', foresrt_model.score(val_X,val_y))

#R^2 (coefficient of determination) regression score function.
#Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). A constant model that always 
print('r2_score', metrics.r2_score(val_y, flightdelay_pred))


```
